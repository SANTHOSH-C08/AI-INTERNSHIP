```python
for i in df.columns:
  if df[i].dtypes !="object":
    sns.boxplot(df[i])
    plt.show()
```

```python
col=['latitude','longitude','age_first_funding_year','age_last_funding_year','relationships','funding_rounds','funding_total_usd','milestones','is_NY','is_MA','is_TX','is_otherstate','is_software','is_web','is_mobile','is_enterprise','is_advertising','is_gamesvideo','is_ecommerce','is_biotech','is_consulting','has_roundC','has_roundD','avg_participants','is_top500']
```

```python
for i in col:
  q1=df[i].quantile(0.25)
  q3=df[i].quantile(0.75)
  iqr=q3-q1
  lf=q1-1.5*iqr
  uf=q3+1.5*iqr
  df=df[(lf <= df[i]) & (uf >= df[i])]
```

```python
for i in df.columns:
    plt.figure(figsize = (12,6))
    sns.boxplot(df[i])
    plt.title(f"checkig outliers for {i}")
    plt.xlabel(i)
    plt.ylabel("Count")
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
  if df[i].dtypes=='object':
    df[i]=le.fit_transform(df[i])
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df["status"]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Performance of my model:", accuracy_score(y_test, y_pred)*100)
```
