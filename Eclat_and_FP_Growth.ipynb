```python
transaction=[]
for i in range(0,len(df)):
  transaction.append([str(df.values[i,j])for j in range(0,20)if str(df.values[i,j])!='nan'])
```

```python
transaction_df = pd.DataFrame(transaction)
```

```python
from pyECLAT import ECLAT

eclat_instance = ECLAT(data=transaction_df)

indices, supports = eclat_instance.fit(
      min_support=0.003,
      min_combination=2,
      max_combination=2
  )
```

```python
eclat_df = pd.DataFrame({
    'Itemset': list(supports.keys()),
    'Support': list(supports.values())
})

eclat_df
```

```python
eclat_df=eclat_df.sort_values(by='Support',ascending=False)
eclat_df
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(transaction).transform(transaction)

df_encoded = pd.DataFrame(
    te_array,
    columns=te.columns_
)
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules

# One-hot encoded transaction data
freq_items = fpgrowth(
    df_encoded,
    min_support=0.003,
    use_colnames=True
)
freq_items
```

```python
rules = association_rules(
    freq_items,
    metric="confidence",
    min_threshold=0.3
)

rules = rules[rules['lift'] >= 2]
```

```python
fp_df = pd.DataFrame({
    'LHS': rules['antecedents'].apply(lambda x: ', '.join(list(x))),
    'RHS': rules['consequents'].apply(lambda x: ', '.join(list(x))),
    'Support': rules['support'],
    'Confidence': rules['confidence'],
    'Lift': rules['lift']
})

fp_df
```

```python
fp_df=fp_df.sort_values(by='Lift',ascending=False)
fp_df
```
