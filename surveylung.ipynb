```python
df['AGE'].fillna(df['AGE'].mean(),inplace=True)
df['ANXIETY'].fillna(df['ANXIETY'].mean(),inplace=True)
df['GENDER'].fillna(df['GENDER'].mode()[0],inplace=True)
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy='median')
df[['AGE']] = imp.fit_transform(df[['AGE']])
df
```

```python
sns.boxplot(df['AGE'])
```

```python
Q1=df['AGE'].quantile(0.25)
Q3=df['AGE'].quantile(0.25)
IQR = Q3-Q1
lower_bound =Q1-1.5*IQR
upper_bound =Q3+1.5*IQR
outliers=df[(df['AGE']<lower_bound) | (df['AGE']>upper_bound)]
outliers
```

```python
from scipy.stats import zscore

z_scores = zscore(df['AGE'])

df[np.abs(z_scores) >2.5]
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(contamination=0.1)
df['outlier'] = iso.fit_predict(df[['AGE','SMOKING']])

df[df['outlier'] == -1]
```

```python
for i in df.columns:
  sns.boxplot(df[i])
  plt.xlabel('count')
  plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt=PowerTransformer(method="yeo-johnson")
df['AGE']=pt.fit_transform(df[['AGE']])
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
  if df[i].dtypes == "object":
    df[i]=le.fit_transform(df[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,:-1]
y=df['LUNG_CANCER']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
dc.fit(x_train,y_train)
y_pred=dc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy score",accuracy_score(y_test,y_pred)*100)
```

```python
from matplotlib import pyplot as plt
from sklearn import tree
fig=plt.figure(figsize=(10,6))
tree.plot_tree(dc,max_depth=15)
plt.show()
```
