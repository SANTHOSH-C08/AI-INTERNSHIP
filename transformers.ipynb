```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
classifier("this is my best day in my whole life")
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "My coach teach me to be discipline in cricket",
    candidate_labels =["education", "sports", "technology"]
)
```

```python
pipe = pipeline("text-generation", model="gpt2")
print(pipe("AI is", max_new_tokens=20))
```
