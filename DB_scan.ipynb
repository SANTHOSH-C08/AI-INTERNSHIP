```python
q=df.iloc[:,3:].values
q
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=10,min_samples=8,metric='euclidean')
```

```python
model=db.fit(q)
```

```python
label=model.labels_
```

```python
label
```

```python
for i in range(len(label)):
    if label[i] == -1:
        label[i] = 0      # replace with nearest/best cluster manually
```
